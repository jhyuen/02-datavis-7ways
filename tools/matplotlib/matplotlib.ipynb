```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('../../cars-sample.csv')
fig, ax = plt.subplots(figsize=(10, 8))
colors = {'ford':'green','toyota':'pink','bmw':'red','honda':'green','mercedes':'blue'}
ax.scatter(x = df['Weight'], y = df['MPG'], c=df['Manufacturer'].map(colors), alpha=0.5, s=0.04*df['Weight'])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.xticks(np.arange(2000, 5001, 1000))
plt.yticks(np.arange(10, 50, 10))

plt.show()
```
